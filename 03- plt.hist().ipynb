```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randn(1000)
plt.hist(data , bins = 50 , color= 'gold')
plt.grid(True)
plt.xlabel('points')
plt.title('Histogram')
plt.show()
```

```python
from random import randint
```

```python
population_age = [randint(1,120) for i in range(40)]
bins = [i for i in range(0,101,10)]
```

```python
plt.hist(population_age , bins , histtype= 'bar' , rwidth=.8)
plt.xlabel('age groups')
plt.ylabel('number of people')
plt.title('Histogram')
plt.show()
```
