# import matplotlib.pyplot as plt

```python
import matplotlib.pyplot as plt
```

#### plt.plot()

```python
x = ['python' , 'c++' , 'C#' , 'visual basic' , 'java' , 'c']
y = [100,60,45,23,40,45]
```

```python
plt.title('programming languages')
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
plt.plot(x,y , color = 'red' , marker = 'o' , linestyle = 'dashed' , linewidth = .5 , markersize = 10)
plt.show()
```
