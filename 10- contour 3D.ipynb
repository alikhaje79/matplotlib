```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d
```

```python
def f(x,y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
```

```python
x = np.linspace(-10 , 10 , 90)
y = np.linspace(-10 , 10 , 90)
X , Y = np.meshgrid(x,y)
```

```python
Z = f(X , Y)
```

```python
fig = plt.figure()
ax = axes3d.Axes3D(fig)
ax.contour3D(X , Y , Z ,40 , cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
