# import matplotlib.pyplot as plt

```python
import matplotlib.pyplot as plt
```

```python
x = [0,2,4,6,8,10]
y = [12,9,6,4,7,3]
```

```python
plt.bar(x,y)
bar_name = ('python' , 'c++' , 'php', 'java' , 'matlab' , 'scala')
plt.xticks(x,bar_name)
plt.show()
```

### for example

```python
plt.bar([.25,1.25,2.25,3.25,4.25],[50,40,70,80,20], label = 'BMW' , width = .5)
plt.bar([.75 , 1.75 , 2.75 , 3.75 , 4.75],[80,20,20,50,60], label = 'Audi' , width = .5 , color = 'pink')
plt.legend()
plt.xlabel('Days')
plt.ylabel('Distance(kms)')
plt.title('information')
plt.show()
```
