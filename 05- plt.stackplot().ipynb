# import matplotlib.pyplot as plt

```python
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]
```

```python
plt.plot([],[] , color = 'pink' , label = 'sleeping' , linewidth = 5)
plt.plot([],[] , color = 'w' , label = 'eating' , linewidth = 5)
plt.plot([],[] , color = 'y' , label = 'working' , linewidth = 5)
plt.plot([],[] , color = 'k' , label = 'playing' , linewidth = 5)
plt.stackplot(days,sleeping,eating,working,playing,colors=['pink' , 'w','y','k'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('stack plot')
plt.legend()
plt.show()
```
